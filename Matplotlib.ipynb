2. __BAR PLOT__ - TO SEE CATAGORICAL DISTRIBUSTION

```python
student = {'sheraz':66, 'asim':68,"saif":50}

names = list(student.keys())
values = list(student.values())

plt.bar(names, values)
plt.show()
```

3. __SCATTER PLOT__

```python
import seaborn as sns

df = sns.load_dataset("titanic")
df.head()
```

```python
plt.scatter(x='age',y='fare', data=df )
plt.show
```

```python
plt.scatter(x='age',y='fare', data=df, c="g" )
plt.scatter(x='survived',y='fare', data=df, c='r' )
plt.show
```

```python
# TWO PLOT AT A TIME

plt.subplot(1,2,1)
plt.scatter(x='age',y='fare', data=df, c="g" )

plt.subplot(1,2,2)
plt.scatter(x='survived',y='fare', data=df, c='r' )

plt.show
```

3. __HISTOGRAM__ - USED  FOR NUMERICAL VALUES

```python
plt.hist(x='age', data=df, color='red')
plt.show
```

```python
# BINS

plt.hist(x='age', data=df, color='red', bins=4)
plt.show
```
