# __SEABORN - Data visualization library based on matplotlib__

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df.head()
```

__LINE PLOT__

```python
sns.lineplot(data=df, x="age",  y='survived')

# We can set the sizze of the graph
# we can Set TRANSPARENCY by using ALPHA
```

```python
# Set TRANSPARENCY by using ALPHA

sns.lineplot(data=df, x="age",  y='survived', alpha=0.3)
```

```python
# Hue is used to visualize the data of different categories in one plot
# We can set color of the graph


sns.lineplot(data=df, x="age",  y='survived', hue='who', color='green')
```

__BAR PLOT__

```python
sns.barplot(data=df, x="sex",  y='age', hue='who', color='green')
```

__SCATTER PLOT__

```python
# STYLE - it gives different shapes to the points based on the column given from data
sns.scatterplot(data=df, x="age",  y='fare', style='sex', hue='who')
```

__HISOGRAM/DISTPLOT__

```python
sns.histplot(data=df, x="age", kde=True, bins=5)

# KDE represent the underlying probability distribution
```

```python
sns.displot(data=df, x="who", kde=True)
```

__JOINT PLOT__- Comination of Scatter plot & Histogram

```python
sns.jointplot(data=df, x='age', y='fare', color='red')
```

```python
# using HUE
sns.jointplot(data=df, x='age', y='fare', color='red', hue='sex')
```

```python
# Passing Regression line through the both, Histogram and Scattre plot

sns.jointplot(data=df, x='age', y='fare', kind='reg')
```

__BOX PLOT__

```python
sns.boxplot(data=df, x='pclass', y="age")
```

__BOXEN PLOT__

```python
sns.boxenplot(data=df, x='pclass', y="age")
```

__PAIR PLOT__

```python
sns.pairplot(data=df, hue='survived')
```

__SWARM PLOT__

```python
sns.swarmplot(data=df, x='sex', y="age", hue='who')
```

__LM PLOT__

```python
sns.lmplot(data=df, x='age',y='fare', hue='who')
```

__CAT PLOT__

```python

sns.catplot(data=df, x='pclass', y='fare', hue='embarked', kind='point')
```

__REL PLOT__

```python
sns.relplot(data=df, x='age', y='fare', hue='embarked')
```

__VIOLIN PLOT__

```python
sns.violinplot(data=df, x='pclass', y="age")
```
